```python
import pandas as pd
```

```python
data = pd.read_csv('../data/data_noncoding.nctools-predictions.eqtls.txt', sep='\t')
```

```python
data['#Chrom'] = data['#Chrom'].astype(str)
```

```python
data.shape
```

```python
deepsea = pd.read_csv('../data/nctools_predictions/deepsea_jobs/infile.vcf.out.logfoldchange')
```

```python
deepsea
```

```python
deepsea['#Chrom'] = deepsea.chr.apply(lambda x: x[3:])
```

```python
deepsea = deepsea.rename(columns={'pos': 'Pos', 'ref': 'Ref', 'alt': 'Alt'})
```

```python
deepsea
```

```python
data.shape, deepsea.shape
```

```python
merged = data.merge(deepsea, how='left', on=['#Chrom', 'Pos', 'Ref', 'Alt'])
```

```python
merged.shape
```

```python
merged.to_csv('../data/data_model/data_noncoding.nctools-predictions.eqtls.deepsea_features.txt', sep='\t', index=False)
```
